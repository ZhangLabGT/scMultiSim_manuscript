```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sns.set(style="darkgrid")
```

```python
res = pd.read_csv(f"./results/grn.csv")
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(6, 4))
res = pd.read_csv(f"./results/nogr.csv")
a = sns.boxplot(res, y="value", x="name", ax=axs[0])
sns.stripplot(res, y="value", x="name", ax=a, alpha=0.5, size=3, dodge=True)
a.set(xlabel='Without GRN', ylabel='Cosine correlation', ylim=(0, 0.9))
a.set_xticklabels(a.get_xticklabels(),rotation = 30, horizontalalignment='right')

res = pd.read_csv(f"./results/grn.csv")
a = sns.boxplot(res, y="value", x="name", ax=axs[1])
sns.stripplot(res, y="value", x="name", ax=a, alpha=0.5, size=3, dodge=True)
a.set(xlabel='With GRN', ylabel='', ylim=(0, 0.9))
a.set_xticklabels(a.get_xticklabels(),rotation = 30, horizontalalignment='right')

plt.savefig("velo.pdf")
```
